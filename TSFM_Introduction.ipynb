### I. Data Processing

```python
df_mo.columns
```

```python
# Drop and agg monthly dataset

# Select EST Med Route only
df_mo = df_mo[df_mo["Route"] == "Eastern Mediterranean Route"]

# Drop Border and Nation
df_mo = df_mo.drop(columns=["Border type or inland", "Nationality"])

# Column to row
df_mo = df_mo.melt(id_vars="Route", var_name="Date", value_name="Detections")

# Changing to time format
df_mo["Date"] = pd.to_datetime(df_mo["Date"], format="%b-%y")

# Agg by sum
df_mo = df_mo.groupby("Date")["Detections"].sum().sort_index()

# Freq assigned
df_mo = df_mo.asfreq("MS")

df_mo
```

```python
import matplotlib.pyplot as plt

# Plot the original time series
plt.figure(figsize=(12, 6))
plt.plot(df_mo, label="Original", color="blue")

# Mean and standard deviation
rolling_mean = df_mo.rolling(window=12).mean()
rolling_std = df_mo.rolling(window=12).std()

# Plot the rolling statistics
plt.plot(rolling_mean, label="12-Month Rolling Mean", color="orange")
plt.plot(rolling_std, label="12-Month Rolling Std", color="green")

plt.title("Eastern Mediterranean Route - Detections Over Time")
plt.xlabel("Date")
plt.ylabel("Detections")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform ADF test
result = adfuller(df_mo.dropna())
print("ADF Statistic:", result[0])
print("p-value:", result[1])
print("Critical Values:")
for key, value in result[4].items():
    print(f"   {key}: {value}")
```

```python
# => Since p-value < 0.05 => The seires is stationary
```

```python
from sktime.forecasting.base import ForecastingHorizon
from sktime.split import temporal_train_test_split

# Last 3 months is y_test since Forecast Horizon is always 3 months
y_train, y_test = temporal_train_test_split(df_mo, test_size=3)
fh = ForecastingHorizon(y_test.index, is_relative=False)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

sarimax_model = SARIMAX(
    y_train,
    order=(2, 1, 1),
    seasonal_order=(1, 1, 1, 4),
    enforce_stationarity=False,
    enforce_invertibility=False
)
sarimax_result = sarimax_model.fit()
sarimax_pred = sarimax_result.get_forecast(steps=3).predicted_mean
```

```python
sarimax_pred
```

### III. Prophet

```python
from sktime.forecasting.fbprophet import Prophet

prophet_model = Prophet(
    seasonality_mode="additive",
    changepoint_prior_scale=0.5,
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False
)

prophet_model.fit(y_train)
prophet_pred = prophet_model.predict(fh)
```

```python
prophet_pred
```

### IV. LSTM


```python
from sktime.forecasting.neuralforecast import NeuralForecastLSTM

model = NeuralForecastLSTM(
    input_size=24,
    max_steps=100
)

model.fit(y_train, fh=fh)
lstm_pred = model.predict(fh=fh)
```

```python
lstm_pred
```

### V. TimeLLMForecaster

```python
from sktime.forecasting.time_llm import TimeLLMForecaster

llm_forecaster = TimeLLMForecaster(
    pred_len=3,
    seq_len=24,
    llm_model='GPT2'
)
llm_forecaster.fit(y_train, fh=fh)
llm_pred = llm_forecaster.predict(fh=fh)
```

```python
llm_pred
```

### VI. Evaluation

```python
from sktime.performance_metrics.forecasting import (
    mean_absolute_error,
    mean_absolute_percentage_error,
    mean_squared_error
)
from sklearn.metrics import r2_score
import numpy as np

preds = {
    "SARIMAX": sarimax_pred,
    "Prophet": prophet_pred,
    "LSTM": lstm_pred,
    "TimeLLM": llm_pred
}

# Store results
results = {}

for name, pred in preds.items():
    mse = mean_squared_error(y_test, pred)
    results[name] = {
        "MAE": mean_absolute_error(y_test, pred),
        "MAPE": mean_absolute_percentage_error(y_test, pred),
        "RMSE": np.sqrt(mse),
        "R2": r2_score(y_test, pred)
    }
```

```python
results_df = pd.DataFrame(results).T
print(results_df)
```
